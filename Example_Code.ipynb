```python
# @title Introduction
# This notebook demonstrates three basic operations in Google Colab:
# 1. Downloading a satellite image from the BMKG website.
# 2. Saving the downloaded image into Google Drive.
# 3. Displaying the image inside the notebook.

print("Welcome! This notebook will guide you through downloading and displaying a satellite image.")
```

```python
# @title 1. Download Satellite Image
# We will download a sample Himawari satellite image using wget.
# After running this cell, you should see the PNG image saved in the Colab workspace.

!wget https://inderaja.bmkg.go.id/IMAGE/HIMA/H08_EH_Indonesia.png

print("Download completed.")
```

```python
# @title 2. Save Image to Google Drive
# In this step, we mount Google Drive and copy the downloaded file into a folder.
# This allows the file to persist, even after the Colab session ends.

from google.colab import drive
drive.mount('/content/drive')

import os

save_path = "/content/drive/MyDrive/Colab_Satellite_Images"
os.makedirs(save_path, exist_ok=True)

!cp H08_EH_Indonesia.png /content/drive/MyDrive/Colab_Satellite_Images/

print("Image successfully saved to Google Drive.")
```

```python
# @title 3. Display the Image
# Here we display the downloaded satellite image using IPython.display.Image.
# This allows the image to appear directly inside the output cell.

from IPython.display import Image

print("Displaying image from local file:")
Image("H08_EH_Indonesia.png")
```
